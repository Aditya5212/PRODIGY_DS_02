### Outliers cleaning


```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Pclass',y='Age',data=data,palette='winter')
```

```python
def imp_age(row):
    age = row['Age']
    pclass = row['Pclass']
    if pd.isnull(age):
        if pclass == 1:
            return 32  # Replace with the mean age of Pclass 1
        elif pclass == 2:
            return 28  # Replace with the mean age of Pclass 2
        elif pclass == 3:
            return 25  # Replace with the mean age of Pclass 3
    else:
        return age
```

```python
data['Age'] = data[['Age','Pclass']].apply(imp_age,axis=1)
```

#### Data is cleaned as per our requirement


```python
sns.set_style('whitegrid')
sns.countplot(x='Survived',hue='Pclass',data=data,palette='rainbow')
```
